# **Hybrid BSM Pricer**
First, the math behind the model.

Again, the goals of the pricer are to:
1. Create a Black-Scholes model.
2. The model can learn and correct for residual pricing errors using real market data.
3. No more constant volatility term: the model uses a fused volatility estimate that incorporates both GARCH and an RNN.
4. The model can identify and compensate for sudden discontinuities in price dynamics due to its jump detection module.
